```python
url = "https://opendata.paris.fr/api/records/1.0/search/?dataset=arrondissements&rows=20"
import requests
r = requests.get(url)
json = r.json()
```

```python
json.keys()
```

```python
arr_geo = json['records']
```

```python
len(arr_geo)
```

```python
arr_geo[0]
```

```python
import folium
paris = folium.Map(location = [48.856578, 2.351828], zoom_start = 12)
```

```python
paris
```

```python
arr_geo[0]['fields']['l_ar']
```

```python
arr_geo[0]['fields']['geom_x_y']
```

```python
folium.Marker(arr_geo[0]['fields']['geom_x_y'], popup = arr_geo[0]['fields']['l_ar']).add_to(paris)
paris
```

```python
arr_geo[0]['fields']['geom']['coordinates']
```

```python
coord = arr_geo[0]['fields']['geom']['coordinates'][0]
[x[::-1] for x in coord]
```

```python
folium.PolyLine([x[::-1] for x in coord]).add_to(paris)
paris
```
